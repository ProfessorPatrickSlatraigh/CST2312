###Warning: Infinite Loops    

While there are loops which are designed to repeat execution thousands, or millions of times, there are cases where a loop can seem stuck with no hope of ever ending.  Have you ever had an application hang?  Frozen?  It seemed that everything just stopped working?  That can often be the result of a loop in the application logic which is never evaluating a conditional test successfully.  Perhaps the programmer assumed that some resource or device would be connected to the computer and available to respond to a request but there is no response, and hence no condition to end the loop.  When a loop keeps processing its code block over and over again with no means of stopping, that is referred to as an **infinite loop**.

###Warning: Other Flaws In Condition Evaluation    

Remember that computers do not work well when there is ambiguity, at least our current computers do not -- quantum computing may result in machines that are better suited to dealing with ambiguity.    

In the meantime, we need to accept that fact that our code will be executed by a machine that takes what we ask it to do literally.  If we set a condition in the `while` statement of a loop and that condition tests whether two variables have equal value, then they must have exactly equal value for the condition to evaluate as **True**.  We may arithmetically calculate that two variables should have the same value but if your program has one variable as an `int` type and the other as a `float` they may not have exactly the same value in the computer's memory and an `==` comparison of them may evaluate to **False**.

Again, we will need to plan and test our logic and code well to avoid such problems.

###Loops: Exception Testing     

Some loops may continue executing their `code block` without counting how many times that `code block` has executed.  This type of program flow is not uncommon in **Industrial Control** (IC) systems.  There are even programs running in a car which rely on this type of program flow.   

Another type of exception processing loop may repeat a step or steps until a condition changes, indicating the end of processing

Of the example use cases we considered earlier, the following would typically rely upon testing an exception or a threshold value and not simply count the number of times their `code block` executed before breaking out of a loop:    

* A light testing for motion WHILE it is dark     
* A cash register printing WHILE there is paper    
* A shopping application asking the user for items WHILE the user is not finished with their order    
* An elevator keeping its doors open WHILE there is too much weight inside    




````python
# Loops: Exception Testing    

ready_for_input = True

while ready_for_input:
    user_input = input('Input (or type Enter to end):')
    print('You input:', user_input)
    ready_for_input = user_input 

print('Thank you for your input')
````

## 6(5).5 Finishing iterations with continue

###Continuing Another Iteration Of The Loop    


* The `continue` statement ends the current iteration of the loops `code block`    

* `continue` jumps to the top of the loop and starts the next iteration    

* A `continue` statement can happen anywhere in the loop.     




````python
while True:
    line = input('> ')
    if line[0] == '#':
        continue
    if line == 'done':
        break
    print(line)
print('Done!')
````

Let's add some user-friendly functionality to that last snippet with a revised code block (below).

````python
# Continuing Another Iteration Of The Loop   

print('Type # to start an entry over.')
print('Enter "done" when finished entries.\n')

while True:
    user_input = input('> ')
    if user_input[0] == '#' :
        continue
    if user_input.lower() == 'done' :
        break
    print(user_input)

print('Done!')
````

### 6(5).7.2 Maximum and minimum loops

````python
largest = None
print('Before:', largest)
for itervar in [3, 41, 12, 9, 74, 15]:
    if largest is None or itervar > largest :
        largest = itervar
    print('Loop:', itervar, largest)
print('Largest:', largest)
````

````python
smallest = None
print('Before:', smallest)
for itervar in [3, 41, 12, 9, 74, 15]:
    if smallest is None or itervar < smallest:
        smallest = itervar
    print('Loop:', itervar, smallest)
print('Smallest:', smallest)
````

### Traversing Lists

Since lists are mutable, it is often desirable to traverse a list, modifying each of its elements. 

The following squares all the numbers from 1 to 5:

````python
numbers = [1, 2, 3, 4, 5]
print("The list numbers: ", numbers) # print the list before processing the for loop
for index in range(len(numbers)):
    numbers[index] = numbers[index]**2
print("The list numbers:", numbers) # print the list after processing the for loop
````

Take a moment to think about **range(len(numbers))** until you understand how it works. 

We are interested here in both the **value** and its **index** within the list, so that we can assign a new value to it.


### Enumerating Lists

Thinking about **range(len(numbers))** and understanding how it works and can be applied to many uses cases in processing lists, we can learn another way to solve the same problem. 

We are interested here in both the **value** and its **index** within the list, so that we can assign a new value to it.

This pattern is common enough that Python provides a nicer way to implement it:

`numbers = [1, 2, 3, 4, 5]`

`for index, value in enumerate(numbers):`

> `numbers[index] = value**2`



````python
numbers = [1, 2, 3, 4, 5]
print("The list numbers: ", numbers) # print the list before processing the for loop
for index, value in enumerate(numbers):
    numbers[index] = value**2
print("The list numbers:", numbers) # print the list after processing the for loop
````

**enumerate** generates both the index and the value associated with it during the list traversal. 

Try this next example to see more clearly how enumerate works:


````python
for index, value in enumerate(['banana', 'apple', 'pear', 'quince']):
    print(index, value)
````

### Join() Function



*note: In the Elkner text, which is written based on Python 2.0, the join() function is referenced as a method imported from the string module -- that is no longer the case in Python 3.0 and beyond.*



As we can see from the last example, str cant be used to join a list of characters together. To do this we could use the **join()** method which works on string objects:


`char_list = list("Frog")`

`char_list`

['F', 'r', 'o', 'g']

`s = ""`

`joined_characters = s.join(char_list, '')`

'Frog'



````python
char_list = list("Frog")

char_list

print("Character list: ", char_list)
````

We create an empty string **s** to apply the join() method to, and we give the **join()** method a list of characters to be joined using the string **s** as a separator.  In this example we want no separator so we create **s** as an empty string.

````python
s = ""
joined_characters = s.join(char_list)
print("Character list: ", char_list)
print("Joined characters: ", joined_characters)
````

If we had wanted the list of characters to be separated by a hyphen, we would create the string **s** with the assignment of a hyphen as 

`s = '-'`

Then apply the join method to the string **s** (the separator), with the list of characters to join as the argument.

````python
s = "-"
joined_characters = s.join(char_list)
print("Character list: ", char_list)
print("Joined characters: ", joined_characters)
````

Suppose we wanted to join together the strings in a list of three digit numbers and have each element from the list separated by a hyphen "-". 

 If our list of three digit numbers was:

`three_digit_numbers =  ["718", "212", "917", "201", "914", "516"]`


````python
three_digit_numbers = ["718", "212", "917", "201", "914", "516"]
s = "-"
joined_number_string = s.join(three_digit_numbers)
print("Three digit number list: ", three_digit_numbers)
print("Joined number string: ", joined_number_string)
````

The variable name of the separator does not have to be **s**. 

In a more tangible example, we may have the four components of an IP address in a list variable **ip_elements** and want to combine them with a dot "**.**" as a separator to have a fully-qualified IPv4 address in a string variable **ip_address**. 

`ip_elements = ["192", "168", "0", "1"]`


````python
ip_elements = ["192", "168", "0", "1"]
s = "."
ip_address = s.join(ip_elements)
print("IPv4 address elements: ", ip_elements)
print("IPv4 address: ", ip_address)
````

### Split() Function
